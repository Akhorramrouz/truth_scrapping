```python
import os
import pandas as pd
def delete_empty_files(min_num=30):
    for f_name in os.listdir():
        if f_name.endswith(".csv"):
            df = pd.read_csv(f_name)
            if len(df)<min_num:
                os.remove(f_name)
                print(f_name, " is deleted")
```
